```python
df = pd.read_csv("Data_Analyst_Data.csv")

# Create a summary table
summary_table = df.groupby(['job_title', 'experience_level']).size().reset_index(name='frequency')
# Pivot the table using the pivot_table function
pivot_table = summary_table.pivot_table(index='job_title', columns='experience_level', values='frequency', aggfunc='sum', fill_value=0)
# Sort by total frequencies to identify the top 10 company_locations
total_frequencies = pivot_table.sum(axis=1)
top_10_job_title_indices = total_frequencies.nlargest(10).index
# Use the top 10 company_locations to filter the pivot table
top_10_job_titles = pivot_table.loc[top_10_job_title_indices]
# Display the top 10 locations pivot table
top_10_job_titles
```

```python
import pandas as pd 

df = pd.read_csv("Data_Analyst_Data.csv")
# Group the data by â€˜job_titleâ€™ and â€˜job_categoryâ€™ and calculate salary statistics
salary_stats = df.groupby(['job_title', 'job_category'])['salary'].agg(['mean', 'max', 'median', 'min']).reset_index()
# Rename columns for clarity
salary_stats.columns = ['Job Title', 'Job Category', 'Average Salary', 'Maximum Salary', 'Median Salary', 'Minimum Salary']
# Display the salary statistics table
salary_stats.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(72, 36))

# Sort the data by average salary
sorted_data = salary_stats.sort_values("Average Salary", ascending=False)

# Plot the average salary for each job title, colored by job category
sns.barplot(x="Average Salary", y="Job Title", data=sorted_data, hue="Job Category", dodge=False)

# Customize the plot
ax.set_xlabel("Average Salary", fontsize=12)
ax.set_ylabel("Job Title", fontsize=12)
ax.set_title("Average Salary by Job Title and Category", fontsize=16)
plt.legend(title="Job Category")

# Show the plot
plt.show()
```

```python
# Filtering the original data frame 'data_df' to select only United States AND Data Analysis AND Full-time
data_us = df[(df['company_location'] == 'United States') & 
                  (df['job_category'] == 'Data Analysis') & 
                  (df['employment_type'] == 'Full-time')]

# Creating a summary table grouped by work setting, company size, and experience level
summary_table = data_us.groupby(['work_setting', 'company_size', 'experience_level'])['salary_in_usd'].mean().round().reset_index(name='average_salary')

# Convert the 'average_salary' to thousands (k) in the summary table
summary_table['average_salary'] = (summary_table['average_salary']).round(1)

# Combining the columns 'company_size' and 'work_setting' into a single column
summary_table['company_setting'] = summary_table['company_size'] + ' - ' + summary_table['work_setting']

# Selecting only the desired columns
summary_table = summary_table[['company_setting', 'experience_level', 'average_salary']]

# Reordering columns for better display in the heatmap
summary_table = summary_table.sort_values(by='average_salary', ascending=False)
summary_table
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pivot the summary_table for the heatmap using keyword arguments
heatmap_data = summary_table.pivot(index="company_setting", columns="experience_level", values="average_salary")

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="coolwarm", linewidths=.5)
plt.title('Average Salary by Company Setting and Experience Level')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='average_salary', y='company_setting', hue='experience_level', data=summary_table)
plt.title('Distribution of Average Salaries by Company Setting and Experience Level')
plt.xlabel('Average Salary (k)')
plt.ylabel('Company Setting')
plt.legend(title='Experience Level')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(x='company_setting', y='average_salary', hue='experience_level', data=summary_table, marker='o')
plt.xticks(rotation=45)
plt.title('Average Salary Trends by Company Setting and Experience Level')
plt.xlabel('Company Setting')
plt.ylabel('Average Salary (k)')
plt.legend(title='Experience Level')
plt.show()
```
